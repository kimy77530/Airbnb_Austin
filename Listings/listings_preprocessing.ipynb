# Listings.csv Preprocessing: Amenities

```python
# Each listing has an amenities column that includes a string of words & phrases that we want to convert to lists
amenities_list = inside_airbnb_listings['amenities'].to_list()
```

```python
# Create a combined amenities list for all unique amenities across all Austin Airbnbs
comb_amenities = []
for i in amenities_list:
    comb_amenities += eval(i)
    
```

```python
# There are too many amenities. Below we find the top 30 amenities across all airbnbs and use those.
from collections import Counter

counts = Counter(comb_amenities)
top30 = pd.Series(counts).sort_values(ascending = False).head(30).index
```

```python
# These are the top 30 most popular amenities across all Austin airbnb listings
top30
```

```python
# Establish a new Amenities column and add to the dataframe
inside_airbnb_listings['amenities2'] = inside_airbnb_listings['amenities'].apply(lambda x: eval(x))
inside_airbnb_listings['amenities2']
```

```python
# Create a column for the dataframe with either a '1' or '0' of if the listing has a smoke alarm 
# Based on the amenities tab

inside_airbnb_listings['smoke alarm'] = inside_airbnb_listings['amenities2'].apply(lambda x: int('Smoke alarm' in x))
# inside_airbnb_listings
```

```python
# Doing the same as above but now for all the amenities in the top20 list that was created

for i in top30:
    inside_airbnb_listings[i] = inside_airbnb_listings['amenities2'].apply(lambda x: int(i in x))

# Note: Check if this makes a duplicate column for smoke alarm 0/1
```

```python
# Add a column for how many of the top30 amenities each listing has
inside_airbnb_listings['Number of Amenities'] = inside_airbnb_listings['amenities2'].str.len()
```

# Listings.csv Preprocessing: Locational Distance Feature Addons

```python
'''
In this section, a locational-based feature is added on. Using the longitude and latitude of each airbnb provided, 
the distance from locations of interest is added for each listing. 
Locations of interest include the Texas Capitol, ... MORE TO BE ADDED
Their locations are sourced from GoogleMaps and then the distance from each listing to the locations of interest
are calculated to be then added on as a feature column for analysis. 
'''
```

```python
# Make latitude^2 and longitude^2 columns for distance-based feature

inside_air_listing_final['latitudesq'] = inside_air_listing_final['latitude'] * inside_air_listing_final['latitude']
inside_air_listing_final['longitudesq'] = inside_air_listing_final['longitude'] * inside_air_listing_final['longitude']
```

```python
from geopy.distance import geodesic
```

```python
# Coordinates sourced from GoogleMaps - lat, long

# Texas Capitol coords 
t_cap = (30.27494739275662, -97.74038005499503)

# Barton Spring coords 
barton_springs = (30.264313040205653, -97.77130428958358)

# Lyndon B. Johnson Library & Museum
lyndon_library = (30.285932343275405, -97.72918194747368)

# Bullock State Museum 
bullock = (30.280420360552085, -97.73904724562665)

# Ann W. Richards Congress Bridge
congress_bridge = (30.262679277470813, -97.74476043398026)

# Zilker Metro Park
zilker = (30.26742874371506, -97.772938709499)

# University of Texas Tower
ut_tower = (30.28639177754446, -97.73941402320536)

# DKR Stadium UT Football
dkr_stadium = (30.283675939587837, -97.73240276590948)

# Mexic-Arte Museum
mex_arte = (30.26697300377302, -97.74270657261515)

# McKinney State Park
mckinney = (30.193317823399394, -97.72431871269195)

# Umlauf Sculpture Garden and Museum
umlauf = (30.263499208682216, -97.76621997630959)

# Mount Bonnell
bonnell = (30.321843891816012, -97.77289219659404)

# Museum of the Weird
museum_weird = (30.267176641146776, -97.73873437630948)

# Sourced these 'top options' from https://www.planetware.com/tourist-attractions-/austin-us-tx-a.htm
```

```python
# Add a column including the listing's distance to the Texas Capitol in KM
 
inside_air_listing_final['t_cap'] = inside_air_listing_final.apply(lambda z: geodesic(t_cap, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['barton_springs'] = inside_air_listing_final.apply(lambda z: geodesic(barton_springs, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['bullock'] = inside_air_listing_final.apply(lambda z: geodesic(bullock, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['congress_bridge'] = inside_air_listing_final.apply(lambda z: geodesic(congress_bridge, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['zilker'] = inside_air_listing_final.apply(lambda z: geodesic(zilker, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['ut_tower'] = inside_air_listing_final.apply(lambda z: geodesic(ut_tower, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['dkr_stadium'] = inside_air_listing_final.apply(lambda z: geodesic(dkr_stadium, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['mex_arte'] = inside_air_listing_final.apply(lambda z: geodesic(mex_arte, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['mckinney'] = inside_air_listing_final.apply(lambda z: geodesic(mckinney, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['umlauf'] = inside_air_listing_final.apply(lambda z: geodesic(umlauf, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['bonnell'] = inside_air_listing_final.apply(lambda z: geodesic(bonnell, (z.latitude, z.longitude)).km, axis=1)
inside_air_listing_final['museum_weird'] = inside_air_listing_final.apply(lambda z: geodesic(museum_weird, (z.latitude, z.longitude)).km, axis=1)
```

```python
# Review scores for location - fix NaNs. Nearest Neighbors on 10 closest locations
# review_scores_location
from tqdm import tqdm

inside_air_listing_final['review_score_location_original'] = inside_air_listing_final['review_scores_location']
null_df_scores = inside_air_listing_final.loc[inside_air_listing_final['review_scores_location'].isnull()]

for i, location in tqdm(null_df_scores.iterrows(), total=null_df_scores.shape[0]):
    temp_dist_col = inside_air_listing_final.apply(lambda z: geodesic((location.latitude, location.longitude), (z.latitude, z.longitude)).km, axis=1)
    inside_air_listing_final_copy = inside_air_listing_final.copy()
    inside_air_listing_final_copy['temp_dist_col'] = temp_dist_col
    inside_air_listing_final_copy.sort_values('temp_dist_col', inplace=True) # default ascending
    temp_mean_val = inside_air_listing_final_copy['review_score_location_original'].dropna().reset_index()[:10].mean()

    inside_air_listing_final.loc[i, 'review_scores_location'] = temp_mean_val.iloc[1]
    

# Takes about 2 - 3 hours to run. Exported to .csv for ease.
```
